```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor, Pool
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
concated = pd.read_csv('../data/onehote.csv')
```

```python
concated.fillna(0, inplace=True)
```

```python
X = concated[:200000].drop(['money', 'patient_ID'], axis=1)
y = concated[:200000]['money']
```

```python
test = concated[200000:]
```

```python
X_test = test.drop(['money', 'patient_ID'], axis=1)
y_test = test['money']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X.astype('int'), y, random_state=1234)
```

```python
TrainPool = Pool(X_train, y_train)
EvalPool = Pool(X_val, y_val)
```

```python
reg = CatBoostRegressor(learning_rate=0.1, iterations=1000, depth = 8, use_best_model=True)
```

```python
reg.fit(TrainPool, eval_set=EvalPool)
```

```python
reg.score(X_test, y_test)
```
